```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
info_cmpl = pd.read_pickle('../data/raw/morris_sample_results_N10_lev4_corrected.pkl')
```

### **Configurations table**

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
info_cmpl.iloc[[1,16,40,120,122,179,180.197]]
```

```python
indices = [1, 16, 40, 120, 122, 179, 180, 197]
plt.figure(figsize=(10, 6))  # Optional: sets the figure size

for i in indices:
    x = info_cmpl['ifc'].values[i]
    y = info_cmpl['Ucell'].values[i]
    plt.plot(x, y, label=f'Index {i}')

plt.axhline(y=0, color='gray', linestyle='--', linewidth=1)
plt.xlabel('ifc')   # Label for x-axis
plt.ylabel('Ucell') # Label for y-axis
plt.title('Ucell vs ifc for selected indices')
plt.legend()
plt.grid(True)
plt.show()
```
